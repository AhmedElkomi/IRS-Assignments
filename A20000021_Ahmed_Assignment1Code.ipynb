```python
import pandas as pd
import numpy as np

data = pd.read_csv('amazon_data_cleaned.csv')

print(data.head())

# Create a list of user IDs
user_ids = [f'User_{i+1}' for i in range(5)]

# Create a user-item matrix with random ratings (1 to 5) for available products
# Reindex the product IDs if necessary for matrix creation
product_ids = data['Product']

# Create a DataFrame with users as rows and products as columns
user_item_matrix = pd.DataFrame(
    np.random.randint(1, 6, size=(len(user_ids), len(product_ids))),
    index=user_ids,
    columns=product_ids
)

# Save the user-item matrix to CSV if you need to use it later
user_item_matrix.to_csv('user_item_matrix.csv', index=True)

import os

desktop_path = os.path.join(os.path.expanduser("~"), "Desktop", "user_item_matrix.csv")

user_item_matrix.to_csv(desktop_path, index=True)

print(f"File saved to: {desktop_path}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from scipy.stats import pearsonr

# User-Based Cosine similarity
user_cosine_similarity = cosine_similarity(user_item_matrix)
user_cosine_similarity_df = pd.DataFrame(user_cosine_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)

# Item-Based Cosine similarity
item_cosine_similarity = cosine_similarity(user_item_matrix.T)
item_cosine_similarity_df = pd.DataFrame(item_cosine_similarity, index=user_item_matrix.columns, columns=user_item_matrix.columns)

# Pearson correlation similarity
def pearson_similarity(matrix):
    # Creating an empty DataFrame to store the similarities
    similarity_matrix = pd.DataFrame(index=matrix.index, columns=matrix.index)
    for user1 in matrix.index:
        for user2 in matrix.index:
            if user1 != user2:
                # Calculate Pearson correlation
                valid_ratings = matrix.loc[[user1, user2]].dropna(axis=1, how='any')
                if valid_ratings.shape[1] > 1:  # Ensure there's more than one rating
                    similarity_matrix.loc[user1, user2], _ = pearsonr(valid_ratings.loc[user1], valid_ratings.loc[user2])
            else:
                similarity_matrix.loc[user1, user2] = 1  # Diagonal elements should be 1
    return similarity_matrix.fillna(0)

# User-Based Pearson Similarity
user_pearson_similarity_df = pearson_similarity(user_item_matrix)

# Item-Based Pearson Similarity
item_pearson_similarity_df = pearson_similarity(user_item_matrix.T)
```

```python
# Function to predict ratings using collaborative filtering
def predict_ratings(matrix, similarity, kind='user', top_n=5):
    """
    Predict ratings based on collaborative filtering.
    :param matrix: User-item matrix
    :param similarity: Similarity matrix
    :param kind: 'user' or 'item' based collaborative filtering
    :param top_n: Number of top items to recommend
    :return: DataFrame with predicted ratings and top-N recommendations
    """
    # Initialize prediction matrix with zeros
    prediction = np.zeros(matrix.shape)
    
    # User-based collaborative filtering
    if kind == 'user':
        for i in range(matrix.shape[0]):
            # Calculate weighted sum of ratings
            sim_scores = similarity[i, :]
            sim_scores_sum = np.abs(sim_scores).sum()  # Sum of similarities
            
            # Prediction for each item by user i
            prediction[i, :] = sim_scores.dot(matrix) / (sim_scores_sum + 1e-9)  # Avoid division by zero

    # Item-based collaborative filtering
    elif kind == 'item':
        for j in range(matrix.shape[1]):
            # Calculate weighted sum of ratings
            sim_scores = similarity[j, :]
            sim_scores_sum = np.abs(sim_scores).sum()  # Sum of similarities
            
            # Prediction for each user for item j
            prediction[:, j] = matrix.dot(sim_scores) / (sim_scores_sum + 1e-9)  # Avoid division by zero

    # Convert to DataFrame
    predicted_ratings_df = pd.DataFrame(prediction, index=matrix.index, columns=matrix.columns)
    
    # Create top-N recommendations
    top_n_recommendations = predicted_ratings_df.apply(lambda row: row.nlargest(top_n).index.tolist(), axis=1)
    return predicted_ratings_df, top_n_recommendations
```

```python
# Predict ratings using User-based Cosine Similarity
user_cosine_pred, user_cosine_top_n = predict_ratings(user_item_matrix, user_cosine_similarity_df.values, kind='user')

# Predict ratings using Item-based Cosine Similarity
item_cosine_pred, item_cosine_top_n = predict_ratings(user_item_matrix, item_cosine_similarity_df.values, kind='item')

# Predict ratings using User-based Pearson Correlation
user_pearson_pred, user_pearson_top_n = predict_ratings(user_item_matrix, user_pearson_similarity_df.values, kind='user')

# Predict ratings using Item-based Pearson Correlation
item_pearson_pred, item_pearson_top_n = predict_ratings(user_item_matrix, item_pearson_similarity_df.values, kind='item')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

def calculate_accuracy(predicted_ratings, true_ratings):
    # Flatten the arrays for easier comparison
    predicted_flat = predicted_ratings.values.flatten()
    true_flat = true_ratings.values.flatten()
    
    # Calculate RMSE and MAE
    rmse = mean_squared_error(true_flat, predicted_flat, squared=False)
    mae = mean_absolute_error(true_flat, predicted_flat)

    return rmse, mae

# Example of calculating accuracy for User-based CF (Cosine)
rmse_user_cosine, mae_user_cosine = calculate_accuracy(user_cosine_pred, user_item_matrix)
print(f'User-based CF (Cosine) - RMSE: {rmse_user_cosine}, MAE: {mae_user_cosine}')

# Repeat for other prediction methods
rmse_user_pearson, mae_user_pearson = calculate_accuracy(user_pearson_pred, user_item_matrix)
print(f'User-based CF (Pearson) - RMSE: {rmse_user_pearson}, MAE: {mae_user_pearson}')

rmse_item_cosine, mae_item_cosine = calculate_accuracy(item_cosine_pred, user_item_matrix)
print(f'Item-based CF (Cosine) - RMSE: {rmse_item_cosine}, MAE: {mae_item_cosine}')

rmse_item_pearson, mae_item_pearson = calculate_accuracy(item_pearson_pred, user_item_matrix)
print(f'Item-based CF (Pearson) - RMSE: {rmse_item_pearson}, MAE: {mae_item_pearson}')
```

The User-Based Collaborative Filtering with Pearson correlation performs best with the lowest RMSE (1.05) and MAE (0.84), indicating better accuracy in predicting user ratings compared to the other models. Lower RMSE and MAE values suggest fewer errors in predictions, making this approach more reliable for recommending products.
